```python
import pandas as pd

titanic = pd.read_csv("train.csv")

titanic.head()
```

```python
sobreviventes = titanic["Survived"].value_counts()

print("Morreram:", sobreviventes[0])
print("Sobreviveram:", sobreviventes[1])
```

```python
mulheres = titanic[titanic["Sex"] == "female"]
homens = titanic[titanic["Sex"] == "male"]

print("Mulheres que sobreviveram:", mulheres["Survived"].mean() * 100, "%")
print("Homens que sobreviveram:", homens["Survived"].mean() * 100, "%")
```

```python
import seaborn as sns

sns.barplot(x="Sex", y="Survived", data=titanic)
```
